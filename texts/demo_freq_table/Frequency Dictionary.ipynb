```python
import pandas as pd
```

```python
f = open("book.txt", "r")
book = f.read()
```

```python
words = book.split(' ')
word_dict = dict.fromkeys(words, 0)

for word in words:
    word_dict[word]+=1
```

```python
word_dict
```

```python
pd.DataFrame(word_dict, index=[0]).to_csv('sleepy_hollow.csv')
```
